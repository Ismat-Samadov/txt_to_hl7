```python
def convert_to_int(value):

    try:
        return int(float(value))
    except (ValueError, TypeError):
        return value

pd.set_option('display.max_rows', None)

def get_value(field, row):

      source_column = lookup_table[field]["Source Column"]
      default_value = lookup_table[field]["Default Value"]

      if source_column:
          if isinstance(source_column, list):
              return "_".join([str(row.get(col, '')) for col in source_column])
          return row.get(source_column, default_value)
      return default_value
```

```python
def parse_to_hl7(df_input, lookup_table):

    hl7_lines = []
    obx_counter = {}
    obr_counter = {}
    empty_list = []
    count = 0

    for index, row in df_input.iterrows():

      PatientMRN= row['PatientMRN']
      if PatientMRN not in obr_counter:
            obr_counter[PatientMRN] = 1
      else:
            obr_counter[PatientMRN] += 1

      RecordedDTTM= row['RecordedDTTM']
      if RecordedDTTM not in obx_counter:
            obx_counter[RecordedDTTM] = 1
      else:
            obx_counter[RecordedDTTM] += 1

      #Create MSH segment
      dt_object = datetime.strptime(row['PerformedDTTM'], "%Y-%m-%d %H:%M:%S.%f")

      # Format the datetime object to the desired string format
      converted_datetime = dt_object.strftime("%Y%m%d%H%M")
      converted_datetime2 = dt_object.strftime("%Y%m%d%H%M%S")

      uniqueid = f"{row['PatientMRN']}_{row['PatientID']}_{converted_datetime}" 
      address = f"{row['APM_Patient_Street1']}^{row['APM_Patient_Street1']}^{row['APM_Patient_Street2']}^{row['APM_Patient_City']}^{row['APM_Patient_State']}^{row['APM_Patient_Zip_Code']}" 

        
      msh_segment = [
          f"MSH|^~\&|{get_value('MSH-3', row)}|{get_value('MSH-4', row)}|{get_value('MSH-5', row)}|{get_value('MSH-6', row)}|{get_value('MSH-7', row)}||{get_value('MSH-9', row)}| {uniqueid} |{get_value('MSH-11', row)}|{get_value('MSH-12', row)}"

              ]

      # Create PID segment
      pid_segment = [
                  f"PID|||{get_value('PID-3.1', row)}^^^^{get_value('PID-3.5', row)}||{get_value('PID-5.1', row)}^{get_value('PID-5.2', row)}||{get_value('PID-7', row)}|{address}"

                  ]
      # Create OBR segment
      obr_segment = [

                  f"OBR|{obr_counter[PatientMRN]}||||||{converted_datetime2}"

              ]
        
      obx_segment = [
            f"OBX|{obx_counter[RecordedDTTM]}||{row.get('LoincLabCode', row)}^{row.get('Metric_Abbr', row)}^LN||{row.get('NumericFinding', row)}|{row.get('Unit_Abbr', row)}"
        ]

      hl7_message = "\n".join(msh_segment + pid_segment + obr_segment + obx_segment)

      hl7_lines.append(hl7_message)

    return hl7_lines
# Parse the input DataFrame to HL7 formatted output
```

```python
hl7_output = parse_to_hl7(df_input, lookup_table)
# hl7_output
```

```python
def sanitize_filename(filename):
    sanitized = re.sub(r'[<>:"/\\|?*]', '_', filename)
    sanitized = sanitized.replace(' ', '_')
    return sanitized


def remove_duplicate_segments(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    unique_segments = {}
    output_lines = []
    current_pid = None
    include_pid = False

    for line in lines:
        parts = line.split('|')
        segment_type = parts[0]
        
        if segment_type == 'MSH':
            if len(parts) > 10:
                unique_id = parts[9].strip()
                key = f"MSH_{unique_id}"
                
                if key not in unique_segments:
                    unique_segments[key] = True
                    output_lines.append(line)
                    include_pid = True  # Flag to include the next PID segment
        
        elif segment_type == 'PID':
            if include_pid or current_pid is None:
                current_pid = line
                output_lines.append(line)
                include_pid = False
        
        elif segment_type == 'OBR':

            if len(parts) > 7:
                parts[1] = '1'  # Replace the second element with '1'
                line = '|'.join(parts)
                observation_date = parts[7]
                key = f"OBR_{observation_date}"
                
                if key not in unique_segments:
                    unique_segments[key] = True
                    output_lines.append(line)
        
        else:
            # Keep all other segments (like OBX)
            output_lines.append(line)

    with open(file_path, 'w') as file:
        file.writelines(output_lines)

    return file



file_paths = []
for index, row in df_input.iterrows():
    filename = f"{row['PatientMRN']}"
    sanitized_filename = sanitize_filename(filename)
    file_path = f"{sanitized_filename}.hl7"
    file_paths.append(file_path)
    with open(file_path, 'w') as file:
        for line in hl7_output:
                file.write(line + "\n")

for file_path in set(file_paths):
    remove_duplicate_segments(file_path)



# print(hl7_output)
```
