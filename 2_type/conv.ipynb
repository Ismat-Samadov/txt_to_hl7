```python
import csv

def create_hl7_message(content_file):
    hl7_message = ""

    # Open and parse the content file
    with open(content_file, 'r') as cf:
        reader = csv.reader(cf, delimiter='|')
        for row_num, row in enumerate(reader, start=1):
            # Skip rows with zero columns (empty rows)
            if len(row) == 0:
                print(f"Warning: Row {row_num} is empty.")
                continue

            # Check if row has enough fields before attempting to access them
            if len(row) < 63:
                print(f"Warning: Row {row_num} has fewer columns than expected: {len(row)} columns.")
                continue  # Skip rows with insufficient data

            try:
                # PID Segment for Patient Information
                patient_id = row[0] if row[0] else "UNKNOWN"
                first_name = row[55] if len(row) > 55 and row[55] else "UNKNOWN"
                last_name = row[54] if len(row) > 54 and row[54] else "UNKNOWN"
                dob = row[64].replace('-', '') if len(row) > 64 and row[64] else ""
                sex = row[63] if len(row) > 63 and row[63] else ""
                address = f"{row[56]}^^{row[58]}^{row[59]}^{row[60]}^{row[61]}" if len(row) > 61 and any(row[56:62]) else ""
                phone = row[65] if len(row) > 65 and row[65] else ""

                # Construct the PID segment, skipping empty address and phone fields
                pid_segment = f"PID|1||{patient_id}^^^MRN||{last_name}^{first_name}^^^||{dob}|{sex}"
                if address:
                    pid_segment += f"|||{address}"
                if phone:
                    pid_segment += f"||{phone}"
                pid_segment += "\n"
                hl7_message += pid_segment

                # AL1 Segment for Allergy Information
                allergy_name = row[4] if len(row) > 4 and row[4] else "UNKNOWN"
                allergy_code = row[8] if len(row) > 8 and row[8] else "UNKNOWN"
                status = row[6] if len(row) > 6 and row[6] else "UNKNOWN"
                category = row[5] if len(row) > 5 and row[5] else "UNKNOWN"
                reaction_date = row[14] if len(row) > 14 and row[14] else ""

                hl7_message += f"AL1|1||{allergy_name}^{allergy_code}^RXNORM|{category}|{status}|{row[26] if len(row) > 26 else ''}|{reaction_date}\n"

                # OBX Segments for each Reaction and Severity
                for i in range(1, 6):
                    reaction = row[27 + (i - 1)] if len(row) > 27 + (i - 1) else None
                    severity = row[32 + (i - 1)] if len(row) > 32 + (i - 1) else None
                    if reaction and reaction != "No Reactions" and severity and severity != "No Severity":
                        hl7_message += f"OBX|{i}|ST|Reaction{i}^{reaction}^SNOMED||{reaction}|{severity}|||F|||{reaction_date}\n"

            except IndexError as e:
                print(f"Error processing row {row_num}: {e}")
                continue

    return hl7_message

# Example usage
content_file = 'ALG.txt'  # Replace with your actual file path
hl7_message = create_hl7_message(content_file)

# Output HL7 message
print(hl7_message)

# Save HL7 message to a file
with open('allergy_message.hl7', 'w') as f:
    f.write(hl7_message)
```
