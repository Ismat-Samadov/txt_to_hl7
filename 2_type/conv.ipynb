```python
import json
from datetime import datetime

# Utility functions
def str_to_lower(s):
    return s.lower()

def str_trim(s):
    return s.strip()

def date_change_format(date_str, from_format, to_format):
    if not date_str.strip():  # If the date string is empty or only contains whitespace
        return ""  # Return an empty string or handle accordingly
    return datetime.strptime(date_str, from_format).strftime(to_format)

def CurrentDateTime(include_time=False):
    if include_time:
        return datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    else:
        return datetime.now().strftime("%Y-%m-%d")

# Set the encoded value in HL7 field
def setEncodedValue(field, value):
    field.value = value

# HL7 Message structure (simplified for this example)
class HL7Message:
    def __init__(self):
        self.MSH = Segment()
        self.PID = Segment()
        self.Allergens = []

class Segment:
    def __init__(self):
        self.SendingApplication = Field()
        self.SendingFacility = Field()
        self.ReceivingApplication = Field()
        self.ReceivingFacility = Field()
        self.DateTimeOfMessage = Field()
        self.MessageType = MessageType()
        self.MessageControlID = Field()
        self.ProcessingID = Field()
        self.VersionID = Field()
        self.PatientIdentifierList = [Field()]
        self.PatientName = [Field()]
        self.AdministrativeSex = Field()
        self.PatientAddress = [Field()]
        self.HomePhoneNumber = [Field()]
        self.BusinessPhoneNumber = [Field()]

class Field:
    def __init__(self):
        self.value = ""

class MessageType:
    def __init__(self):
        self.MessageCode = ''
        self.TriggerEvent = ''

# Function to convert pipe-delimited file into structured in_data format
def convert_to_in_data(file_path):
    in_data = {"ALG": []}
    expected_num_fields = 65  # Number of expected fields based on your structure

    # Read the file and process each line
    with open(file_path, 'r') as file:
        for line_number, line in enumerate(file, start=1):
            values = line.strip().split('|')

            # Pad missing fields with empty strings
            if len(values) < expected_num_fields:
                values.extend([''] * (expected_num_fields - len(values)))

            allergy_entry = {
                "APM_Patient_Number": values[0],
                "AllergyID": values[2],
                "AllergenName": values[4],
                "AllergenType": values[5],
                "Status": values[6],
                "ReactionSeverity1": values[29],
                "Reaction1": values[22],
                "RxNormCode": values[44],
                "UNIICode": values[42],
                "RecordedDTTM": values[16],
                "RecordedBYLastName": values[18],
                "RecordedBYFirsttName": values[19],
                "RecordedBYNPI": values[43],
                "Annotation": values[21],
                "ActionTaken": values[15],
                "EstReactDate": values[14],
                "APM_Patient_Last_Name": values[50],
                "APM_Patient_First_Name": values[49],
                "APM_Patient_MI": values[51],
                "APM_Patient_Street1": values[52],
                "APM_Patient_Street2": values[53],
                "APM_Patient_City": values[54],
                "APM_Patient_State": values[55],
                "APM_Patient_Zip_Code": values[56],
                "APM_Patient_HomePhone": values[59],
                "APM_Patient_Email": values[64]
            }
            
            # Add entry to the in_data structure
            in_data["ALG"].append(allergy_entry)

    return in_data

# Initialize out object for HL7 message construction
out = HL7Message()

# Example usage
file_path = 'ALG_2.txt'
in_data = convert_to_in_data(file_path)

# MSH Segment (HL7 Segment)
out.MSH.SendingApplication.value = "SendingApp_TW"  # Example value
out.MSH.SendingFacility.value = "DefaultFacility"
out.MSH.ReceivingApplication.value = "ReceivingApp"
out.MSH.ReceivingFacility.value = "ReceivingFac"

setEncodedValue(out.MSH.DateTimeOfMessage, date_change_format(CurrentDateTime(True), "%Y-%m-%d %H:%M:%S", "%m/%d/%Y %H:%M:%S"))

# PID Segment (HL7 Segment)
out.PID.PatientIdentifierList[0].value = in_data['ALG'][0]['APM_Patient_Number']
out.PID.PatientName[0].FamilyName = in_data['ALG'][0]['APM_Patient_Last_Name']
out.PID.PatientName[0].GivenName = in_data['ALG'][0]['APM_Patient_First_Name']


# IAM Segment - Updated to handle empty dates
for i in range(len(in_data['ALG'])):
    if (in_data['ALG'][i]['Status'] == "Active" and 
        len(in_data['ALG'][i]['AllergenName']) > 0 and 
        "no known" not in str_to_lower(in_data['ALG'][i]['AllergenName'])):

        sIAMseg = f"{str_to_lower(in_data['ALG'][i]['AllergenType']).replace(' ', '')}^{in_data['ALG'][i]['AllergenType']}|"

        if len(str_trim(in_data['ALG'][i]['RxNormCode'])) > 0:
            sIAMseg += f"{str_trim(in_data['ALG'][i]['RxNormCode'])}^{str_trim(in_data['ALG'][i]['AllergenName'])}^RXN|"
        else:
            sIAMseg += f"^{str_trim(in_data['ALG'][i]['AllergenName'])}^NAME|"

        # Reaction Severity and Reactions
        sIAMseg += f"{in_data['ALG'][i]['ReactionSeverity1']}|"
        sReactions = "~".join([f"{in_data['ALG'][i].get(f'Reaction{j}', 'No Reactions')}^" for j in range(1, 6)])

        sIAMseg += f"{sReactions}|A||||NTL001^{str_trim(in_data['ALG'][i]['AllergenName'])}^EXTELG|"

        # Estimated Reaction Date
        est_react_date = str_trim(in_data['ALG'][i]['EstReactDate'])
        if len(est_react_date) > 0 and est_react_date[:4] != "1900":
            sIAMseg += f"{date_change_format(est_react_date[:19], '%Y-%m-%d %H:%M:%S', 'yyyyMMdd')}||"
        else:
            sIAMseg += "||"  # Append empty reaction date

        # Recorded Date (handling empty RecordedDTTM)
        recorded_dttm = in_data['ALG'][i]['RecordedDTTM'][:19]
        if recorded_dttm.strip():
            sIAMseg += f"{date_change_format(recorded_dttm, '%Y-%m-%d %H:%M:%S', 'HL7')}|"
        else:
            sIAMseg += "|"

        sIAMseg += f"{in_data['ALG'][i]['RecordedBYLastName']}^{in_data['ALG'][i]['RecordedBYFirsttName']}||||"

        # Append IAM Segment
        out.Allergens.append(sIAMseg)

# Example: Print the IAM segments
print("IAM Segments:")
for i, allergen in enumerate(out.Allergens, start=1):
    print(f"IAM Segment {i}: {allergen}")
    
```
